## **Graph 1**

Displaying the percentage of how many songs listed in the data sheet are an album, compilation, or single into a pie graph.



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/Spotify_Youtube.csv')

#counts how many album types are listed in the data set
df_album_type_count = df['Album_type'].value_counts()

# Create a pie chart
labels = df_album_type_count.index.tolist()
sizes = df_album_type_count.values.tolist()

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=30)
plt.axis('equal')
plt.title('Album Types')
plt.legend(labels, loc='best')

plt.show()
```

## **Graph 2**

Showcasing the difference in views from youtube and streams from spotify of the top 10 artists


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/Spotify_Youtube.csv')

# Group the songs by artist and views from youtube and streams from spotify
df_artist_group = df.groupby('Artist')[['Views', 'Stream']].sum()

# Sort the artists descending order by the sum of views and streams
df_artist_sort = df_artist_group.sort_values(['Views', 'Stream'], ascending=False)

# Store the top 10 artists with the most number of views on YouTube and streams on Spotify
top_10 = df_artist_sort.head(10)

# Create two separate DataFrames for views and streams
df_views = df.groupby('Artist')['Views'].sum().sort_values(ascending=False)[:10]
df_streams = df.groupby('Artist')['Stream'].sum().sort_values(ascending=False)[:10]

# Set the width of the bars
barWidth = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(df_views))
r2 = [x + barWidth for x in r1]

# Create the figure and axes objects
fig, ax = plt.subplots(figsize=(20,10))

# Create the first set of bars
ax.bar(r1, df_views, color='#EE3224', width=barWidth, label='Views')

# Create the second set of bars
ax.bar(r2, df_streams, color='#F78F1E', width=barWidth, label='Streams')

# Set the x-axis tick marks to be the artist names
ax.set_xticks([r + barWidth / 2 for r in range(len(df_views))])
ax.set_xticklabels(top_10.index)

# Set the title and labels
ax.set_title('Top 10 Artists on YouTube and Spotify')
ax.set_xlabel('Artist')
ax.set_ylabel('Number of Views/Streams')

# Add the legend
ax.legend()

plt.show()
```

## **Graph 3**

Display the loudness calculated from all songs listed in the data set.




```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

df = pd.read_csv('data/Spotify_Youtube.csv')

# Calculate mean value
df_mean_Loudness = np.mean(df['Loudness'])

fig = px.histogram(df, x = 'Loudness',)
fig.update_layout(title ='Distribution of Loudness', title_x=0.5, title_xanchor= 'center',width = 1000, height = 600)
fig.add_annotation(x=0,y=1.05, text=f"Mean: {df_mean_Loudness:.2f}", showarrow=False, xref="paper", yref="paper", font=dict(size=15))
fig.update_xaxes(title_text='Loudness')
fig.update_yaxes(title_text='Tracks')
```

9. Key: the key in which the track is played, with integers mapped to pitches utilizing standard Pitch Class notation
10. Loudness: the overall loudness of the track in decibels (dB)
11. Speechiness: an indicator of whether spoken words are present in a track, with values above 0.66 suggesting that the recording is mostly speech-like
12. Acousticness: a metric from 0.0 to 1.0 representing the level of confidence that the track is acoustic
13. Instrumentalness: a prediction of whether a track contains no vocals, with values approaching 1.0 indicating a higher likelihood of instrumental content
14. Liveness: a measure detecting whether an audience is present in the recording, with higher values indicating a greater probability of the track being performed live
15. Valence: a measure ranging from 0.0 to 1.0 describing the degree of positivity conveyed by the track's music
16. Tempo: an estimate of the track's overall tempo in beats per minute (BPM)
17. Duration_ms: the track's duration in milliseconds
18. Stream: the number of times the song has been streamed on Spotify
19. Url_youtube: the URL link to any associated video on YouTube
Title: the title of the video clip on YouTube
20. Channel: the name of the channel that published the video
21. Views: the number of views the video has received on YouTube
22. Likes: the number of likes the video has received on YouTube
23. Comments: the number of comments the video has received on YouTube
23. Description: a description of the video on YouTube
24. Licensed: an indicator of whether the video represents licensed content
