```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('data/cases_by_age_vac_status.csv')
```

## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python
df_gas = df[(df['Date'].str.contains('2022')) & (df['Fuel Type'] == 'Regular Unleaded Gasoline')]
mean_prices = df_gas.mean()
mean_prices.plot(kind ='bar')
plt.xlabel('City')
plt.ylabel('Mean Price')
plt.title('Mean Regular Unleaded Gasoline Price in Different Cities in 2022')
plt.show()
```

## Question 2

Plot the histogram of the Regular Unleaded Gasoline price in Ottawa for 2022. Choose the appropriate parameters.

```python
ottawa_gas= df_gas[['Date', 'Ottawa']]
ottawa_gas.hist(bins=10, grid=False)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Regular Unleaded Gasoline Price in Ottawa in 2022')
plt.show()
```

## Question 3

Plot the ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.

```python
df = df_gas['Ottawa'].values
x = np.sort(df)
n = x.size
y = np.arange(1, n+1) / n

plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('Price')
plt.ylabel('ECDF')
plt.title('ECDF of Regular Unleaded Gasoline Price in Ottawa in 2022')
plt.show()
```
